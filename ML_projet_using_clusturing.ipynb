```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype == 'object':
            df[col] = df[col].fillna(df[col].mode()[0])
        else:
            df[col] = df[col].fillna(df[col].mean())
```

```python
df.isnull().sum()
```

```python
drop_columns = [
    "user_id","churned"
]

df = df.drop(columns=[col for col in drop_columns if col in df.columns])
```

```python
df["Age_group"]=pd.cut(
    df["age"],
    bins=[0,19,30,60,100],labels=["teen","adult","old","too old"])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_cols = ['gender', 'country', 'subscription_type', 'payment_method', 'primary_device', 'favorite_genre']

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
features=["account_age_months","subscription_type","country",'age','monthly_fee','primary_device','devices_used']
for c in features[1:]:
  plt.figure(figsize=(4,3))
  sns.boxplot(y=df[c]);plt.title(c);plt.show()
for c in features[1:]:
  q1,q3=df[c].quantile([0.25,0.75]);iqr=q3-q1
  low=q1-1.5*iqr;high=q3+1.5*iqr
  df=df[(df[c]>=low)&(df[c]<=high)]
X=df[features]
print(df.shape)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import Birch
```

```python
from sklearn.cluster import KMeans
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from cycler import K
wcss=[]
for k in range(1,11):
  km=KMeans(n_clusters=k,random_state=42,n_init=10)
  km.fit(X_scaled)
  wcss.append(km.inertia_)
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel('K');plt.ylabel('WCSS');plt.grid();plt.show()
```

```python
kmeans=KMeans(n_clusters=2,random_state=42,n_init=10)
df["Cluster"]=kmeans.fit_predict(X_scaled)
print('Silhouette:',silhouette_score(X_scaled,df["Cluster"]))
```

```python
display(df.groupby('Cluster')[features[1:]].mean())
print(df['Cluster'].value_counts())
```

```python
for k in range(2,11):
  kmeans=KMeans(n_clusters=k,random_state=42,n_init=10)
  labels=kmeans.fit_predict(X_scaled)
  score=silhouette_score(X_scaled,labels)
  print(f"Silhouette score for k= {k} is equal score is ={score}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
cluster_names = {0: 'Not Churn', 1: 'Churn'}
df['Cluster_Label'] = df['Cluster'].map(cluster_names)
plt.figure(figsize=(10, 7))
sns.scatterplot(x='account_age_months', y='monthly_fee', hue='Cluster_Label', data=df, palette='viridis', s=50, alpha=0.6)
plt.title('K-Means Clustering (K=2): Churn vs. Not Churn')
plt.xlabel('Account Age (Months)')
plt.ylabel('Monthly Fee')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Customer Segment')
plt.show()
```
